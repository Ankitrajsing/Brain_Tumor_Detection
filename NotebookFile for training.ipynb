```python
import os
import json
from glob import glob

# Define input folder (where JSON files are stored)
json_folder = "/content/drive/MyDrive/Dataset new/Json (LabelMe)"

# Define output folder for YOLO .txt labels
output_folder = "/content/drive/MyDrive/Dataset new/Labels/Tumor"

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Get all JSON files in the folder
json_files = glob(os.path.join(json_folder, "*.json"))

# Function to convert LabelMe JSON to YOLO segmentation format with normalization
def convert_json_to_yolo(json_file, output_folder):
    try:
        with open(json_file, "r") as file:
            data = json.load(file)

        # Extract image filename and dimensions
        image_filename = os.path.basename(data["imagePath"]).replace("\\", "/")
        image_width = data["imageWidth"]
        image_height = data["imageHeight"]
        txt_filename = os.path.splitext(image_filename)[0] + ".txt"

        # Process all annotated objects
        annotations = []
        for shape in data["shapes"]:
            class_label = 0  # Assuming single class "tumor", update if necessary
            points = shape["points"]

            # Normalize points to [0,1] range with 6 decimal precision
            normalized_points = [(round(x / image_width, 6), round(y / image_height, 6)) for x, y in points]

            # Convert to YOLO format: <class_id> x1 y1 x2 y2 ... xn yn
            flat_points = [str(coord) for point in normalized_points for coord in point]
            annotation_line = f"{class_label} " + " ".join(flat_points)
            annotations.append(annotation_line)

        # Save to YOLO .txt file
        txt_output_path = os.path.join(output_folder, txt_filename)
        with open(txt_output_path, "w") as txt_file:
            txt_file.write("\n".join(annotations))

        print(f"✅ Converted: {json_file} → {txt_output_path}")

    except KeyError as e:
        print(f"🚨 KeyError: Missing key {e} in {json_file}, skipping.")
    except FileNotFoundError:
        print(f"🚨 FileNotFoundError: Skipping {json_file} because the file or folder was not found.")
    except Exception as e:
        print(f"❌ Error processing {json_file}: {e}")

# Convert all JSON files
for json_file in json_files:
    convert_json_to_yolo(json_file, output_folder)

print("✅ Conversion completed for all JSON files!")
```

```python
import os
import cv2

# Define paths
image_folder = "/content/drive/MyDrive/Dataset new/Images/Healthy"  # Update with actual dataset path
output_folder = "/content/drive/MyDrive/Dataset new/Label/Healthy"  # Output folder for labels

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Supported image formats
image_extensions = (".jpeg")

# Get all image files in the folder
image_files = [f for f in os.listdir(image_folder) if f.lower().endswith(image_extensions)]

# Process each image
for image_file in image_files:
    image_path = os.path.join(image_folder, image_file)
    image = cv2.imread(image_path)

    if image is None:
        print(f"Skipping {image_file}, unable to read.")
        continue

    height, width, _ = image.shape  # Get image dimensions

    # Define a polygon covering the full image with more than 5 points
    polygon = [
        0, 0,                    # Top-left corner
        width // 2, 0,            # Top-center
        width - 1, 0,            # Top-right corner
        width - 1, height // 2,  # Middle-right
        width - 1, height - 1,   # Bottom-right corner
        width // 2, height - 1,   # Bottom-center
        0, height - 1,           # Bottom-left corner
        0, height // 2           # Middle-left
    ]

    # Normalize polygon coordinates to [0,1] range
    normalized_polygon = [round(coord / width, 6) if i % 2 == 0 else round(coord / height, 6)
                          for i, coord in enumerate(polygon)]

    # Convert to string format
    polygon_str = " ".join(map(str, normalized_polygon))

    # Generate label file content
    label_content = f"1 {polygon_str}"  # Class ID 1

    # Save the label file
    label_filename = os.path.splitext(image_file)[0] + ".txt"
    label_path = os.path.join(output_folder, label_filename)

    with open(label_path, "w") as label_file:
        label_file.write(label_content)

    print(f"Generated label: {label_path}")

print("✅ Label generation completed!")
```

```python
import os
import shutil
import random

# ✅ Define paths (Update these)
healthy_image_dir = "/content/drive/MyDrive/Dataset new/Images/Healthy"
healthy_label_dir = "/content/drive/MyDrive/Dataset new/Label/Healthy"
tumor_image_dir = "/content/drive/MyDrive/Dataset new/Images/Tumor"
tumor_label_dir = "/content/drive/MyDrive/Dataset new/Label/Tumor"
output_dir = "/content/drive/MyDrive/Dataset new/Dataset"

# ✅ Create output directories for train, val, test
for split in ["train", "val", "test"]:
    os.makedirs(os.path.join(output_dir, split, "images"), exist_ok=True)
    os.makedirs(os.path.join(output_dir, split, "labels"), exist_ok=True)

# ✅ Function to split dataset
def split_dataset(image_dir, label_dir):
    image_files = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]
    random.shuffle(image_files)

    num_images = len(image_files)
    train_cutoff = int(num_images * 0.7)
    val_cutoff = train_cutoff + int(num_images * 0.2)

    train_files = image_files[:train_cutoff]
    val_files = image_files[train_cutoff:val_cutoff]
    test_files = image_files[val_cutoff:]

    return train_files, val_files, test_files

# ✅ Function to move images & labels
def move_files(file_list, image_dir, label_dir, split):
    for img_file in file_list:
        img_path = os.path.join(image_dir, img_file)
        label_path = os.path.join(label_dir, os.path.splitext(img_file)[0] + ".txt")  # Match label name

        # Move image
        shutil.copy(img_path, os.path.join(output_dir, split, "images", img_file))

        # Move label if exists
        if os.path.exists(label_path):
            shutil.copy(label_path, os.path.join(output_dir, split, "labels", os.path.splitext(img_file)[0] + ".txt"))

# ✅ Split and move files for both classes
healthy_train, healthy_val, healthy_test = split_dataset(healthy_image_dir, healthy_label_dir)
tumor_train, tumor_val, tumor_test = split_dataset(tumor_image_dir, tumor_label_dir)

move_files(healthy_train, healthy_image_dir, healthy_label_dir, "train")
move_files(healthy_val, healthy_image_dir, healthy_label_dir, "val")
move_files(healthy_test, healthy_image_dir, healthy_label_dir, "test")

move_files(tumor_train, tumor_image_dir, tumor_label_dir, "train")
move_files(tumor_val, tumor_image_dir, tumor_label_dir, "val")
move_files(tumor_test, tumor_image_dir, tumor_label_dir, "test")

print("✅ Balanced dataset split completed!")
```

```python
import os
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Paths to labels
gt_path = '/content/drive/MyDrive/Brain Tumor Detection/Dataset/test/labels'
pred_path = '/content/drive/MyDrive/Brain Tumor Detection/yolov12/runs/segment/tumor-seg-pred/labels'

y_true = []
y_pred = []

def get_class_list(filepath):
    if not os.path.exists(filepath):
        return []
    with open(filepath, 'r') as f:
        return [int(line.strip().split()[0]) for line in f.readlines()]

for file in os.listdir(gt_path):
    if not file.endswith('.txt'): continue

    gt_file = os.path.join(gt_path, file)
    pred_file = os.path.join(pred_path, file)

    gt_classes = get_class_list(gt_file)
    pred_classes = get_class_list(pred_file)

    max_len = max(len(gt_classes), len(pred_classes))

    for i in range(max_len):
        true_cls = gt_classes[i] if i < len(gt_classes) else 1  # 1 = Healthy
        pred_cls = pred_classes[i] if i < len(pred_classes) else 1
        y_true.append(true_cls)
        y_pred.append(pred_cls)

# Confusion Matrix
labels = ['Tumor', 'Healthy']
cm = confusion_matrix(y_true, y_pred, labels=[0, 1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix: Tumor vs Healthy (YOLOv12l-seg)")
plt.show()
```
